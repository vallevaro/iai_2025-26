The Traveling Salesman Problem (TSP) is a classic optimization problem where the objective is to find the shortest possible route that visits a set of cities and returns to the origin city. 
This implementation utilizes the **Simulated Annealing** algorithm to find an approximate solution for TSP.

## ‚öôÔ∏è **Algorithm Overview**
Simulated Annealing is inspired by the annealing process in metallurgy. It works by exploring the solution space and accepting worse solutions with a certain probability, allowing it to escape local minima. The algorithm gradually reduces the temperature, which controls the probability of accepting worse solutions.
